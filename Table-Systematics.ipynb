```python
percent_ = False
precision_ = 5
debug_ = False
# labels_samples_ = {
#     "GGToWW-AQGC-A0W1e-6": r"$\gamma\gamma \to WW \quad a_0^W/\Lambda^2 = 1 \times 10^{-6} \,\rm{GeV}^{-2}$",
#     "GGToWW-AQGC-A0W2e-6": r"$\gamma\gamma \to WW \quad a_0^W/\Lambda^2 = 2 \times 10^{-6} \,\rm{GeV}^{-2}$",
#     "GGToWW-AQGC-A0W5e-6": r"$\gamma\gamma \to WW \quad a_0^W/\Lambda^2 = 5 \times 10^{-6} \,\rm{GeV}^{-2}$"
# }
labels_samples_ = {
    "GGToWW-AQGC-A0W5e-7": r"$a_0^W/\Lambda^2 = 5 \times 10^{-7} \,\rm{GeV}^{-2}$",
    "GGToWW-AQGC-A0W1e-6": r"$a_0^W/\Lambda^2 = 1 \times 10^{-6} \,\rm{GeV}^{-2}$",
    "GGToWW-AQGC-A0W2e-6": r"$a_0^W/\Lambda^2 = 2 \times 10^{-6} \,\rm{GeV}^{-2}$",
    "GGToWW-AQGC-A0W5e-6": r"$a_0^W/\Lambda^2 = 5 \times 10^{-6} \,\rm{GeV}^{-2}$"
}
labels_samples_[ "GGToWW-AQGC-muon-A0W5e-7" ] = labels_samples_[ "GGToWW-AQGC-A0W5e-7" ]
labels_samples_[ "GGToWW-AQGC-muon-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-muon-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-muon-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W5e-6" ]
labels_samples_[ "GGToWW-AQGC-electron-A0W5e-7" ] = labels_samples_[ "GGToWW-AQGC-A0W5e-7" ]
labels_samples_[ "GGToWW-AQGC-electron-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-electron-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-electron-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-A0W5e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-muon-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-muon-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-muon-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W5e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-muon-A0W5e-7" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W5e-7" ]
labels_samples_[ "GGToWW-AQGC-2018-muon-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-muon-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-muon-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-muon-A0W5e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-electron-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-electron-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-2017-electron-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W5e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-electron-A0W5e-7" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W5e-7" ]
labels_samples_[ "GGToWW-AQGC-2018-electron-A0W1e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W1e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-electron-A0W2e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W2e-6" ]
labels_samples_[ "GGToWW-AQGC-2018-electron-A0W5e-6" ] = labels_samples_[ "GGToWW-AQGC-electron-A0W5e-6" ]

labels_samples_[ "Background" ] = "Background"
index_names_ = {
    'central': 'Central value',
    'stat': 'Stat.',
    'jes': 'Jet energy scale',
    'jer': 'Jer energy resolution',
    'sf_muon_id': 'Muon Id',
    'sf_electron_id': 'Electron Id',
    'proton_scale': 'Proton momentum scale',
    'proton_eff': 'Proton efficiency',
    'total': 'Total'
}
df__ = df_sum_weights
if percent_: df__ = df__.iloc[ 1:, : ].where( ~( df__.iloc[ 1:, : ] != -1. ) , df__.iloc[ 1:, : ].divide( df__.iloc[ 0 ] ) * 100. )
if percent_: precision_ = 1
columns_ = df__.columns.values
var_name_sf_lepton_id__ = ""
if lepton_type == 'muon':
    var_name_sf_lepton_id__ = 'sf_muon_id'
elif lepton_type == 'electron':
    var_name_sf_lepton_id__ = 'sf_electron_id'
systematics_list_ = [ 'jes', 'jer', var_name_sf_lepton_id__, 'proton_scale', 'proton_eff', 'total' ]
line_str = r"\begin{table}"; print ( line_str )
line_str = r"\begin{{tabular}}{{l | {}}}".format( 'r ' * len( columns_ ) ); print ( line_str )
line_str = " & "
for idx_, label_sample__ in enumerate( columns_ ):
    line_str += labels_samples_[ label_sample__ ]
    if idx_ < ( len( columns_ ) - 1 ): line_str += " & "
line_str += r" \\ "
print ( line_str )
if not percent_:
    for label__ in [ 'central' ]:
        line_str = "{:32}".format( index_names_[ label__ ] )
        line_str += " & "
        vals__ = df__.loc[ label__ ]
        for idx_, label_sample__ in enumerate( columns_ ):
            val__ = vals__.loc[ label_sample__ ]
            if debug_: print ( label__, label_sample__, val__ )
            line_str += "{:20}".format( r"{{:.{:d}f}}".format( precision_ ).format( val__ ) )
            if idx_ < ( len( columns_ ) - 1 ): line_str += " & "
        line_str += r" \\ "
        print ( line_str )   
for label__ in [ 'stat' ]:
    line_str = "{:32}".format( index_names_[ label__ ] )
    line_str += " & "
    vals__ = df__.loc[ label__ ]
    for idx_, label_sample__ in enumerate( columns_ ):
        val__ = vals__.loc[ label_sample__ ]
        if debug_: print ( label__, label_sample__, val__ )
        if not percent_:
            line_str += "{:20}".format( r"$\pm${{:.{:d}f}}".format( precision_ ).format( val__ ) )
        else:
            line_str += "{:20}".format( r"$\pm${{:.{:d}f}}\%".format( precision_ ).format( val__ ) )
        if idx_ < ( len( columns_ ) - 1 ): line_str += " & "
    line_str += r" \\ "
    print ( line_str )
for label_syst__ in systematics_list_:
    line_str = "{:32}".format( index_names_[ label_syst__ ] )
    line_str += " & "
    vals_up_label__ = 'syst_up_' + label_syst__
    vals_dw_label__ = 'syst_dw_' + label_syst__
    vals_up__ = df__.loc[ vals_up_label__ ]
    vals_dw__ = df__.loc[ vals_dw_label__ ]
    for idx_, label_sample__ in enumerate( columns_ ):
        if label_sample__ != 'Background':
            val_up__ = vals_up__.loc[ label_sample__ ]
            val_dw__ = vals_dw__.loc[ label_sample__ ]
            if debug_: print ( label_syst__, label_sample__, val_up__, val_dw__)
            if not percent_:
                if np.abs( val_up__ ) == np.abs( val_dw__ ):
                    line_str += "{:20}".format( r"$\pm${}".format( r"{{:.{:d}f}}".format( precision_ ).format( val_up__ ) ) )
                else:
                    line_str += "{:20}".format( r"$^{{{}}}_{{{}}}$".format( r"{{:+.{:d}f}}".format( precision_ ).format( val_up__ ),
                                                                            r"{{:-.{:d}f}}".format( precision_ ).format( val_dw__ ) ) )
            else:
                if np.abs( val_up__ ) == np.abs( val_dw__ ):
                    line_str += "{:20}".format( r"$\pm${}\%".format( r"{{:.{:d}f}}".format( precision_ ).format( val_up__ ) ) )
                else:
                    line_str += "{:20}".format( r"$^{{{}}}_{{{}}}$\%".format( r"{{:+.{:d}f}}".format( precision_ ).format( val_up__ ),
                                                                              r"{{:-.{:d}f}}".format( precision_ ).format( val_dw__ ) ) )
        else:
            line_str += "{:20}".format( "--" )
        if idx_ < ( len( columns_ ) - 1 ): line_str += " & "
    line_str += r" \\ "
    print ( line_str )
line_str = r"\end{tabular}"; print ( line_str )
line_str = r"\end{table}"; print ( line_str )
```
