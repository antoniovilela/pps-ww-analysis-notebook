```python
norm_to_1fb = True

# categories = [ 'muon-2017', 'electron-2017', 'muon-2018', 'electron-2018' ]
categories = [ 'muon-2018', 'electron-2018' ]
```

```python
df_sum_weights = {}
file_str_ = "table_systematics"
if norm_to_1fb:
    file_str_ += "-norm_to_1fb"
for category__ in categories:
    file_path_ = ""
    if category__ == 'muon-2017':
        file_path_ = file_str_ + "-2017-muon" + ".h5"
    elif category__ == 'electron-2017':
        file_path_ = file_str_ + "-2017-electron" + ".h5"
    elif category__ == 'muon-2018':
        file_path_ = file_str_ + "-2018-muon" + ".h5"
    elif category__ == 'electron-2018':
        file_path_ = file_str_ + "-2018-electron" + ".h5"

    with pd.HDFStore( file_path_, 'r' ) as store_:
        print ( list( store_ ) )
        df_sum_weights[ category__ ] = store_[ "table_systematics" ]
```

```python
df_uncertainties_relative = {}
for category__ in categories:
    df__ = df_sum_weights[ category__ ]
    df_uncertainties_relative[ category__ ] = df__.iloc[ 1:, : ].where( ~( df__.iloc[ 1:, : ] != -1. ) , df__.iloc[ 1:, : ].divide( df__.iloc[ 0 ] ) )
```

```python
import os

map_labels_ = {}
map_labels_[ "GGToWW-AQGC-A0W1e-6" ] = "sig-2017-muon-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-A0W2e-6" ] = "sig-2017-muon-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-A0W5e-6" ] = "sig-2017-muon-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-muon-A0W1e-6" ] = "sig-2017-muon-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-muon-A0W2e-6" ] = "sig-2017-muon-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-muon-A0W5e-6" ] = "sig-2017-muon-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-2017-muon-A0W1e-6" ] = "sig-2017-muon-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-2017-muon-A0W2e-6" ] = "sig-2017-muon-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-2017-muon-A0W5e-6" ] = "sig-2017-muon-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-electron-A0W1e-6" ] = "sig-2017-electron-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-electron-A0W2e-6" ] = "sig-2017-electron-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-electron-A0W5e-6" ] = "sig-2017-electron-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-2017-electron-A0W1e-6" ] = "sig-2017-electron-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-2017-electron-A0W2e-6" ] = "sig-2017-electron-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-2017-electron-A0W5e-6" ] = "sig-2017-electron-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-2018-muon-A0W5e-7" ] = "sig-2018-muon-AQGC-A0W5e-7"
map_labels_[ "GGToWW-AQGC-2018-muon-A0W1e-6" ] = "sig-2018-muon-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-2018-muon-A0W2e-6" ] = "sig-2018-muon-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-2018-muon-A0W5e-6" ] = "sig-2018-muon-AQGC-A0W5e-6"
map_labels_[ "GGToWW-AQGC-2018-electron-A0W5e-7" ] = "sig-2018-electron-AQGC-A0W5e-7"
map_labels_[ "GGToWW-AQGC-2018-electron-A0W1e-6" ] = "sig-2018-electron-AQGC-A0W1e-6"
map_labels_[ "GGToWW-AQGC-2018-electron-A0W2e-6" ] = "sig-2018-electron-AQGC-A0W2e-6"
map_labels_[ "GGToWW-AQGC-2018-electron-A0W5e-6" ] = "sig-2018-electron-AQGC-A0W5e-6"

year_category_ = {}
year_category_[ 'muon-2017' ] = '2017'
year_category_[ 'electron-2017' ] = '2017'
year_category_[ 'muon-2018' ] = '2018'
year_category_[ 'electron-2018' ] = '2018'
lepton_type_category_ = {}
lepton_type_category_[ 'muon-2017' ] = 'muon'
lepton_type_category_[ 'electron-2017' ] = 'electron'
lepton_type_category_[ 'muon-2018' ] = 'muon'
lepton_type_category_[ 'electron-2018' ] = 'electron'
lumi_unc_category_ = {}
lumi_unc_category_[ 'muon-2017' ] = 1.023
lumi_unc_category_[ 'electron-2017' ] = 1.023
lumi_unc_category_[ 'muon-2018' ] = 1.025
lumi_unc_category_[ 'electron-2018' ] = 1.025

for category__ in categories:
    n_obs__ = 0
    n_bkg__ = df_sum_weights[ category__ ][ "Background" ][ "central" ]
    unc_bkg__ = ( df_uncertainties_relative[ category__ ][ "Background" ][ "stat" ] )
    
    var_name_lepton_id__ = ""
    if lepton_type_category_[ category__ ] == 'muon':
        var_name_lepton_id__ = "sf_muon_id"
    elif lepton_type_category_[ category__ ] == 'electron':
        var_name_lepton_id__ = "sf_electron_id"
    
    lumi_unc__ = lumi_unc_category_[ category__ ]
    
    uncertainties__ = [ 'jes', 'jer', var_name_lepton_id__, 'proton_scale', 'proton_eff' ]
    unc_names__ = {}
    unc_names__[ 'jes' ] = '{}_{}'.format( 'jes', year_category_[ category__ ] )
    unc_names__[ 'jer' ] = '{}_{}'.format( 'jer', year_category_[ category__ ] )
    unc_names__[ var_name_lepton_id__ ] = '{}_{}'.format( var_name_lepton_id__, year_category_[ category__ ] )
    unc_names__[ 'proton_scale' ] = '{}_{}'.format( 'proton_scale', year_category_[ category__ ] )
    unc_names__[ 'proton_eff' ] = '{}_{}'.format( 'proton_eff', year_category_[ category__ ] )
#     unc_names__[ 'lumi' ] = 'lumi'
    unc_names__[ 'lumi' ] = '{}_{}'.format( 'lumi', year_category_[ category__ ] )
    unc_names__[ 'signal_stat' ] = '{}_{}_{}'.format( 'signal_stat', lepton_type_category_[ category__ ], year_category_[ category__ ] )
    unc_names__[ 'bkg_norm' ] = '{}_{}_{}'.format( 'bkg_norm', lepton_type_category_[ category__ ], year_category_[ category__ ] )
    
    for column__ in df_sum_weights[ category__ ].columns.values[ : -1 ]:
        label__ = map_labels_[ column__ ]
        
        n_pred__ = df_sum_weights[ category__ ][ column__ ][ "central" ]
        unc_stat__ = ( df_uncertainties_relative[ category__ ][ column__ ][ "stat" ] )
    
        file_str_ = 'datacard_counting'
        if norm_to_1fb:
            file_str_ += '_norm_to_1fb'
        file_path_ = '{}_{}_{}.txt'.format( file_str_, category__, label__ )
        print ( file_path_ )
        dc = open( file_path_, 'w' )

        dc.write('imax *  number of channels\n')
        dc.write('jmax *  number of processes -1\n')
        dc.write('kmax *  number of nuisance parameters (sources of systematical uncertainties)\n')
        dc.write('--------------------------------------------------------------------------------------------------------------------------- \n')

        dc.write( 'bin         signal_region\n' )
        dc.write( 'observation {}\n'.format( n_obs__ ) )

        dc.write('--------------------------------------------------------------------------------------------------------------------------- \n')
        dc.write('--------------------------------------------------------------------------------------------------------------------------- \n')

        dc.write( '{:<30} {:<30} {:<16}\n'.format( 'bin', 'signal_region', 'signal_region' ) )
        dc.write( '{:<30} {:<30} {:<16}\n'.format( 'process', 'signal', 'background' ) )
        dc.write( '{:<30} {:<30} {:<16}\n'.format( 'process', 0, 1 ) )
        dc.write( '{:<30} {:<30.3f} {:<16.3f}\n'.format( 'rate', n_pred__, n_bkg__ ) )

        dc.write('--------------------------------------------------------------------------------------------------------------------------- \n')
        dc.write('--------------------------------------------------------------------------------------------------------------------------- \n')

        for unc__ in uncertainties__:
            label_unc__ = "syst_up_{}".format( unc__ )
            dc.write( '{:<26} {:<3} {:<30.4f} {:<16}\n'.format( unc_names__[ unc__ ], 'lnN', 
                                                                ( 1. + df_uncertainties_relative[ category__ ][ column__ ][ label_unc__ ] ),
                                                                '-' ) )
        
        dc.write( '{:<26} {:<3} {:<30.4f} {:<16}\n'.format( unc_names__[ 'lumi' ], 'lnN', lumi_unc__, '-' ) )
        dc.write( '{:<26} {:<3} {:<30.4f} {:<16}\n'.format( unc_names__[ 'signal_stat' ], 'lnN', 1. + unc_stat__, '-' ) )
        dc.write( '{:<26} {:<3} {:<30} {:<16.4f}\n'.format( unc_names__[ 'bkg_norm' ], 'lnN', '-', 1. + unc_bkg__ ) )
        
        dc.close()
```
